# Q 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
#  connecting webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
#Oppening url
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

# maximize window
driver.maximize_window()

# Scrapping rank
rank=[]
data= driver.find_elements(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td')
for i in data[0:300:6]:
    rank.append(i.text)
    
# Scrapping video name
name=[]
for i in data[1:300:6]:
    name.append(i.text)
    
    
# Scrapping artist name
artist=[]
for i in data[2:300:6]:
    artist.append(i.text)
    
# Scrapping upload date
date=[]
for i in data[4:300:6]:
    date.append(i.text)
    
# Scrapping views
views=[]
for i in data[3:300:6]:
    views.append(i.text)    
    
df=pd.DataFrame({'Rank':rank, 'Video Name':name, 'Artist':artist, 'Upload date':date, 'Views in Billions':views})    
```

st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#  connecting webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
#Oppening url
driver.get("https://www.bcci.tv/")

# maximize window
driver.maximize_window()
time.sleep(5)

search=driver.find_element(By.XPATH,"/html/body/nav/div[1]/div[2]/ul[1]/li[2]")
search.click()  
    
```

```python

while True:
    try:
        
        element= driver.find_element(By.XPATH,'//*[@id="fixtures"]/div[3]/div[2]/div/button')
        webdriver.ActionChains(driver).move_to_element(element).click(element ).perform()
        time.sleep(2)
    except NoSuchElementException:
        break
```

```python
# Scrapping title
title=[]
t1= driver.find_elements(By.XPATH, '//span[@class="matchOrderText ng-binding ng-scope"]')
for i in t1:
    i=i.text
    i=i.split(' ')[1]
    title.append(i)
    
# Scrapping series    
series=[]
s1= driver.find_elements(By.XPATH, '//span[@class="ng-binding"]')
for i in s1:
    i=i.text
    series.append(i)
    
# Scrapping place of match   
place=[]
p1= driver.find_elements(By.XPATH, '//div[@class="fix-place ng-binding ng-scope"]')
for i in p1:
    i=i.text
    i=i.split('-')[1]
    place.append(i)
    
# Scrapping date of match       
date=[]
d1= driver.find_elements(By.XPATH, '//h5[@class="ng-binding"]')
for i in d1:
    i=i.text
    date.append(i)
    

# Scrapping time of match   
time=[]
t1= driver.find_elements(By.XPATH, '//h5[@class="text-right ng-binding"]')
for i in t1:
    i=i.text
    time.append(i)  
    
# Creating dataframe    
df=pd.DataFrame({'Match title':title, 'Series':series, 'Place':place, 'Date':date, 'Time':time})
```

# 9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
#  connecting webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
#Oppening url
driver.get("https://www.imdb.com/list/ls095964455/")

# maximize window
driver.maximize_window()


#scraping required data
name=[]
for i in driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]'):
    i=i.text.split('(')[0]
    i=i.split('.')[1]
    name.append(i)
    
year=[]
for i in driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]'):
    i=i.text.split('(')[1]
    i=i.split(')')[0]
    year.append(i) 
    
Genre=[i.text for i in driver.find_elements(By.XPATH, '//span[@class="genre"]')]    
time_=[i.text for i in driver.find_elements(By.XPATH, '//span[@class="runtime"]')]
rating=[i.text for i in driver.find_elements(By.XPATH, '//div[@class="ipl-rating-star small"]')]
votes=[i.text for i in driver.find_elements(By.XPATH, '//span[@name="nv"]')]


df=pd.DataFrame({'Name':name,'Year Span':year,'Genre':Genre,'Runtime':time_,'Rating':rating,'Votes':votes})
```
