# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
# connecting webdriver
driver =webdriver.Chrome(r"C:\Users\tara\Downloads\chromedriver_win32C:\Users\tara\Downloads\chromedriver_win32\chromedriver.exe")

# Opening Url
driver.get("https://www.naukri.com/")

#Putting designation
designation = driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Analyst')


#Putting location
location=driver.find_element(By.XPATH, '/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
location.send_keys('Banglore')

# Searching
search = driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()


# Scraping Job title
title= []
t1=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in t1:
    i=i.text
    title.append(i)
    
# Scraping location
loc= []
l1=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in l1:
    i=i.text
    loc.append(i)
    
    
# Scraping Company name
cname= []
name=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in name:
    i=i.text
    cname.append(i)
    
    

# Scraping experience required   
exp= []
exp_r=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in exp_r:
    i=i.text
    exp.append(i)
    
    
#Creating Data frame     
df=pd.DataFrame({'Job Title': title[0:10], 'Location':loc[0:10],'Company Name':cname[0:10],'Experience required':exp[0:10]})
```

# Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. ProductDescription
3. Price

To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url : https://www.flipkart.com/
2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual
4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100 sunglasses.

```python
# connecting webdriver
driver =webdriver.Chrome(r"C:\Users\tara\Downloads\chromedriver_win32C:\Users\tara\Downloads\chromedriver_win32\chromedriver.exe")

# Opening Url
driver.get("https://www.flipkart.com/")

# maximize window
driver.maximize_window()

#Minimising the login page
driver.find_element(By.XPATH, '/html/body/div[2]/div/div/button').click()

# Putting product
product = driver.find_element(By.CLASS_NAME,'_3704LK')
product.send_keys('sunglasses')

# Searching
search = driver.find_element(By.CLASS_NAME,'L0Z3Pu')
search.click()
```

```python
# Scrapping brand name
brand=[]
for page in range(0,3):
    brands= driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    for i in brands:
        brand.append(i.text)
    next_b=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_b.click()
    time.sleep(3)
    
# Scrapping product discription  
product_d=[]
for page in range(0,3):
    product= driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in product:
        product_d.append(i.text)
    next_b=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_b.click()
    time.sleep(3)    
    
# Scrapping price    
price=[]
for page in range(0,3):
    p1= driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in p1:
        price.append(i.text)
    next_b=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_b.click()
    time.sleep(3)
       
```

```python
# Creatin data frame    
df=pd.DataFrame({'Brand': brand[0:100],'Product Dispcription':product_d[0:100], 'Price':price[0:100]}) 
```

# Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:

place=FLIPKART
As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100 reviews

```python
# connecting webdriver
driver =webdriver.Chrome(r"C:\Users\tara\Downloads\chromedriver_win32C:\Users\tara\Downloads\chromedriver_win32\chromedriver.exe")

# Opening Url
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")

# maximize window
driver.maximize_window()
```

```python
# Scrapping rating
rating=[]
for page in range(0,11):
    ratings= driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    for i in ratings:
        rating.append(i.text)
    next_b=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_b.click()
    time.sleep(3)
    
    
# Scrapping review summary
r_summary=[]
for page in range(0,11):
    summary= driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    for i in summary:
        r_summary.append(i.text)
    next_b=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_b.click()
    time.sleep(3)
    

    
# Scrapping full review
full_r=[]
for page in range(0,11):
    review= driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')
    for i in review:
        full_r.append(i.text)
    next_b=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_b.click()
    time.sleep(3)
    
```

```python
#Creating data frame 
df=pd.DataFrame({'Rating':rating, 'Review Summary':r_summary,'Full Review':full_r})
```
